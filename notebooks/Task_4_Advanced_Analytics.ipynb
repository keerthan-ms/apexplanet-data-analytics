# Task 4: Advanced Analytics & Hypothesis Testing 🧪

**Objective:** 
Perform statistical analysis using Independent T-Tests to evaluate marketing channel performance. 

This analysis tests whether the observed differences in **Revenue** and **Acquisition Cost** between our highest-performing channel (Email Marketing) and lowest-performing channel (Paid Advertising) are statistically significant, or merely the result of random variance.

```python
import pandas as pd
from scipy import stats
df = pd.read_csv('cleaned_customer_acquisition_data.csv')
```

```python
from scipy import stats

# Isolate the revenue data for our best and worst performing channels
email_revenue = df[df['channel'] == 'email marketing']['revenue']
paid_revenue = df[df['channel'] == 'paid advertising']['revenue']

# Run an Independent T-Test
t_stat, p_value = stats.ttest_ind(email_revenue, paid_revenue, equal_var=False)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# The business translation
if p_value < 0.05:
    print("Conclusion: The difference is STATISTICALLY SIGNIFICANT. Email marketing truly drives different revenue than Paid Advertising.")
else:
    print("Conclusion: Not significant. The difference might just be random chance.")
```

```python
# Isolate the cost data for our best and worst performing channels
email_cost = df[df['channel'] == 'email marketing']['cost']
paid_cost = df[df['channel'] == 'paid advertising']['cost']

# Run an Independent T-Test on Cost
t_stat_cost, p_value_cost = stats.ttest_ind(email_cost, paid_cost, equal_var=False)

print(f"Cost T-Statistic: {t_stat_cost:.4f}")
print(f"Cost P-Value: {p_value_cost:.4f}")

# The business translation
if p_value_cost < 0.05:
    print("Conclusion: The difference is STATISTICALLY SIGNIFICANT. Paid Advertising is mathematically proven to cost more per customer than Email Marketing.")
else:
    print("Conclusion: Not significant. The acquisition costs are statistically similar.")
```
